```python
# Melihat jumlah nilai kosong di setiap kolom
print("Jumlah Nilai Kosong:")
print(data.isnull().sum())
```

**Nilai Kosong**

Dari analisis nilai kosong, ditemukan beberapa kolom dengan jumlah nilai yang hilang:


*   **children**: Ada 4 nilai kosong. Kolom ini mencatat jumlah anak-anak dalam reservasi
*   **country**: Ada 488 nilai kosong. Kolom ini menunjukkan asal negara tamu berdasarkan kode ISO
*   **agent**: Ada 16,340 nilai kosong. Kolom ini merepresentasikan ID agen perjalanan yang memesan kamar

**Menangani Nilai Kosong**

```python
# Kolom children (4 nilai kosong):
# Solusinya mengganti nilai kosong dengan 0 karena jumlah anak yang tidak tercatat kemungkinan besar adalah 0

data['children'] = data['children'].fillna(0)
```

#**Sistem Prediksi Pembatalan Reservasi**

**Masalah**

Manajemen hotel ingin memahami kemungkinan pembatalan reservasi berdasarkan informasi pemesanan. Dengan prediksi ini, manajemen dapat lebih proaktif dalam menangani pembatalan, misalnya dengan mengatur strategi pengisian ulang kamar atau menawarkan promosi kepada pelanggan tertentu.


**Tujuan**

Membuat model prediksi untuk memprediksi apakah sebuah reservasi akan dibatalkan atau tidak, berdasarkan variabel seperti:
- lead_time (waktu antara reservasi dan kedatangan)
- adults, children, babies (jumlah tamu)
- customer_type, meal, dan distribution_channel (informasi pelanggan dan pemesanan)

**Persiapan Data**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Features
features = ['lead_time', 'stays_in_weekend_nights', 'stays_in_week_nights',
            'adults', 'children', 'babies', 'meal', 'customer_type', 'distribution_channel']

# Mengonversi kolom kategori menjadi data numerik (encoding)
data_encoded = pd.get_dummies(data[features], drop_first=True)

# Target is_canceled (yaitu kolom pembatalan reservasi)
target = data['is_canceled']

# Membagi dataset menjadi data train 80% dan data test 20%
X_train, X_test, y_train, y_test = train_test_split(data_encoded, target, test_size=0.2, random_state=42)
```

**Model Prediksi**

```python
# Menggunakan Random Forest
model = RandomForestClassifier(random_state=42, n_estimators=100, max_depth=10)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

**Evaluasi Model**

```python
# Evaluasi model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

**Analisis Pentingnya Fitur**
